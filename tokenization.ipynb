```python
with open('token.txt','r') as file:
  if file.readable():
    data=file.read()
print(data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(stop_words='english')# it works row by row
res=vectorizer.fit_transform([data])
res
```

```python
import re
# step 1: clean and split
words = re.findall(r'\b[a-zA-Z]+\b', data.lower())
# print("All Words:", words)

# step 2: remove duplicates
unique_words = []
for word in words:
    if word not in unique_words:
        unique_words.append(word)

# print("Unique Words:", unique_words)

# step 3: sort alphabetically
unique_words.sort()
# print("Sorted Words:", unique_words)

# step 4: assign index using dictionary
alpha_index = {}
idx = 0
for word in unique_words:
    alpha_index[word] = idx
    idx += 1

print("Alphabetical Index:", alpha_index)
```

```python
import re

# stopwords list
stop_words = ['i', 'me', 'my', 'we', 'our', 'you', 'your', 'he', 'she', 'it', 'its',
              'they', 'them', 'what', 'which', 'who', 'this', 'that', 'these', 'those',
              'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being',
              'have', 'has', 'had', 'do', 'does', 'did', 'will', 'would',
              'can', 'could', 'should', 'may', 'might', 'shall',
              'a', 'an', 'the', 'and', 'but', 'or', 'so', 'yet',
              'in', 'on', 'at', 'by', 'for', 'with', 'about', 'of', 'to',
              'from', 'up', 'down', 'out', 'off', 'over', 'into', 'through',
              'not', 'no', 'nor', 'as', 'just', 'also', 'than', 'then',
              'if', 'while', 'after', 'before', 'when', 'where', 'how']

# step 1: clean and split
words = re.findall(r'\b[a-zA-Z]+\b', data.lower())
# print("All Words:", words)

# step 2: remove stopwords
filtered_words = []
for word in words:
    if word not in stop_words:
        filtered_words.append(word)

# print("Filtered Words:", filtered_words)

# step 3: assign alphabetical index using dictionary
unique_words = sorted(set(filtered_words))
alpha_index = {}
idx = 0
for word in unique_words:
    alpha_index[word] = idx
    idx += 1

print("Alphabetical Index:", alpha_index)
```
