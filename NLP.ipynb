```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
nltk.download('punkt_tab')
# text = "The boys are running faster and studies are better and they working so hard and clever and they will get more chances if they missed they will loss after finishing their work they are going to their home"
text='caresses flies dies mules denied died agreed owned humbled sized meeting stating siezing itemization sensational traditional reference colonizer plotted'
#step 1: Tokenize (split into words)
tokens = word_tokenize(text)
print(tokens)
#step 2: create stemmer object
stemmer = PorterStemmer()  #stemming method
#step 3: apply stemming
# stemmed_words=[]
# for word in tokens:
#   stemmed_words.append(stemmer.stem(word))
stemmed_word=[stemmer.stem(word) for word in tokens]
print(stemmed_word)
```

**nltk stemming:** Disadvantages
* study-studi
* chance=chanc
* flies=fli
* dies=die
* denied=deni
* humbled=humbl
* stating=state
* sensational=sensat
* traditional=tradit

```python
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize

nltk.download('averaged_perceptron_tagger')

def get_pos(tag):
    if tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

sentence = "He was running faster and better"

tokens = word_tokenize(sentence)
tagged = pos_tag(tokens)

lemmatized = [
    lemmatizer.lemmatize(word, get_pos(pos))
    for word, pos in tagged
]

print(lemmatized)
```
