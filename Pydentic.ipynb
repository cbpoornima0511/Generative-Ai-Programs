##Simple pydentic syntax

```python
from pydantic import BaseModel

class User(BaseModel): #BaseModel is superclass we are inheriting it
  name:str
  age:int
  email:str

data={
    'name':'Poornima',
    'age':23,
    'email':'poornima@gmail.com',
    'number':1436759654
}

user=User(**data)   #data will tagged from user class
print(user.name)
print(user.age)
# print(user.number) #AttributeError bcz it is not present in user class
```

#Proper pydentic model

```python
from pydantic import BaseModel, Field
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import PydanticOutputParser
import os

os.environ["GOOGLE_API_KEY"] = "Your_api_key"


# 1️⃣ Define Schema (Production practice)
class QuestionAnswer(BaseModel):
    question: str = Field(description="User question")
    answer: str = Field(description="Correct answer")
    example: str = Field(description="Real world example")


# 2️⃣ Output Parser
parser = PydanticOutputParser(pydantic_object=QuestionAnswer)


# 3️⃣ Prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI tutor."),
    ("human", "Explain this topic: {topic}\n{format_instructions}")
])


# 4️⃣ Model
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)


# 5️⃣ Chain
chain = prompt | llm | parser


# 6️⃣ Invoke
result = chain.invoke({
    "topic": "Machine Learning",
    "format_instructions": parser.get_format_instructions()
})


print(result)
print(result.answer)
print(result.example)
```
