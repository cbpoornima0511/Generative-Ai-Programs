##**QUESTION AND ANSWER CHATBOT**

```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
os.environ["GOOGLE_API_KEY"] = "AIzaSyCVQi5QMgmY9BTNrv743JTGEznFFIdy5Rg"
# from google.colab import userdata
# userdata.get('GOOGLE_API_KEY')

chat_prompt = ChatPromptTemplate.from_messages([
    ("system",  "You are a helpful assistant.Generate only one Questions"),
    ("human", "Question:{topic}"),
])
ans_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant.Detect the answer is correct or wrong if it is correct give Great,else explain the wrong answer"),
    ("human","Question:{topic}\nUser Answer:{answer}")
])
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)
parser = StrOutputParser()
chat_chain = chat_prompt | llm | parser
ans_chain = ans_prompt | llm | parser

chat_result = chat_chain.invoke({
    "level": "Fresher",
    "topic": "machine learning",
    "Total_no_of_questions":1
})
lst_que=chat_result.split('\n',40)
print(lst_que)
user_answer = input("Enter answer:")
ans_result = ans_chain.invoke({
    "topic": lst_que[0],
    "answer": user_answer
})
print(ans_result)
```
