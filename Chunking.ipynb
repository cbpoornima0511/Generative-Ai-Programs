```python
import numpy as np
import faiss
import spacy
from sentence_transformers import SentenceTransformer
from transformers import pipeline
```

```python
embedding_model = SentenceTransformer(model_name)
embedding_model
```

```python
embedding=embedding_model.encode(chunks)
embedding
```

```python
dim=embedding.shape[1]
dim
```

```python
index = faiss.IndexFlatL2(dim)  #L2-similarity search,fix the dimension
index.add(np.array(embedding))  #all the vector embeddings stored in index like object
print(index)
```

```python
generator = pipeline('text-generation',model='gpt2')
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
# ----------------------------
# STEP 5: RAG Function
# ----------------------------
def rag_query(query, top_k=2):
    docs=nlp(query)
    print(docs,type(docs))
    filter_words=[]
    for token in docs:
      filter_words.append(token.lemma_)
      updated_query=' '.join(filter_words)
    print(updated_query)

    # Convert query to embedding
    query_embedding = embedding_model.encode([updated_query])

    # Search similar docs
    distances, indices = index.search(np.array(query_embedding), top_k)
    print(distances,indices)
    print(indices[0])
    # Retrieve relevant documents
    retrieved_docs = [chunks[i] for i in indices[0]]

    # # Create augmented prompt
    context = " ".join(retrieved_docs)
    prompt = f"Answer the question based on the context.\n\nContext: {context}\n\nQuestion: {query}\nAnswer:"

    # # Generate answer
    result = generator(prompt, max_length=150, num_return_sequences=1)

    return result[0]["generated_text"]


# ----------------------------
# STEP 6: Test Query
# ----------------------------
query="What is Machine Learning?".lower()
response = rag_query(query)
print(response)
```
