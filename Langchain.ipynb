```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
os.environ["GEMINI_API_KEY"] = "Your_api_key"

# ── Step 1: Define the prompt template
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Answer in {language}."),
    ("human",  "{question}"),
])

# ── Step 2: Initialize the FREE Gemini model
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0) #temperature=0 -> to avoid hallucination risk

# ── Step 3: Define the output parser
parser = StrOutputParser()

# ── Step 4: Wire into a pipeline using LCEL
# prompt → llm → parser  (the | operator = pipe)
chain = prompt | llm | parser
chain1 = prompt | llm   #without parser

# ── Step 5: Run the pipeline
result = chain.invoke({
    "language": "kanglish",
    "question": "What is machine learning?"
})
result1 = chain1.invoke({
    "language": "kanglish",
    "question": "What is machine learning?"
})


print(result)

```

###CHATBOT

```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# API key
os.environ["GOOGLE_API_KEY"] = "Your_api_key"

# ── Prompt Template
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Answer in {language}."),
    ("human", "{question}")
])

# ── Gemini Model
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)

# ── Output Parser
parser = StrOutputParser()

# ── Pipeline
chain = prompt | llm | parser

print("🤖 Chatbot started (type 'exit' to stop)\n")

# ── Chat Loop
while True:

    user_question = input("You: ")

    if user_question.lower() == "exit":
        print("Chatbot ended.")
        break

    result = chain.invoke({
        "language": "kanglish",
        "question": user_question
    })

    print("Bot:", result)
```

```python
os.environ["GOOGLE_API_KEY"] = "Your_api_key"

# ── Step 1: Define the prompt template
trans_prompt = ChatPromptTemplate.from_messages([
    ("system",  "You are a helpful assistant.Detect user {question} and Translate to in {language}"),
    ("human", "{question}")
])
summ_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful text summarizer.summarize the text into English"),
    ("human", "{question}")
])

# ── Step 2: Initialize the FREE Gemini model
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0) #temperature=0 -> to avoid hallucination risk

# ── Step 3: Define the output parser
parser = StrOutputParser()

# ── Step 4: Wire into a pipeline using LCEL
# prompt → llm → parser  (the | operator = pipe)
trans_chain = trans_prompt | llm | parser
summ_chain = summ_prompt | llm | parser

# ── Step 5: Run the pipeline
trans_result = trans_chain.invoke({
    "language": "English",
    "question": "namskar ellarigu"
})
summ_result = summ_chain.invoke({
    "question":trans_result
})

print(trans_result)
print(summ_result)
```

```python
os.environ["GOOGLE_API_KEY"] = "Your_api_key"

# ── Step 1: Define the prompt template
trans_prompt = ChatPromptTemplate.from_messages([
    ("system",  "You are a helpful assistant.Detect user {question} and Translate to English"),
    ("human", "{question}")
])
summ_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful text summarizer.summarize the text"),
    ("human", "{question}")
])

# ── Step 2: Initialize the FREE Gemini model
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0) #temperature=0 -> to avoid hallucination risk

# ── Step 3: Define the output parser
parser = StrOutputParser()

# ── Step 4: Wire into a pipeline using LCEL
# prompt → llm → parser  (the | operator = pipe)
trans_chain = trans_prompt | llm | parser
summ_chain = summ_prompt | llm | parser

full_chain = (
    {"translated": trans_chain}
    | summ_chain
)
# ── Step 5: Run the pipeline
trans_result = full_chain.invoke({
    "question": "namskar ellarigu"
})

print(trans_result)
```
